**Ejercicio 1:**
Resuelva un sistema de dos ecuaciones lineales con dos incognitas. Suponga que tienen una única solución.


---


**Solución:**
Consideremos que el sistema es de la forma:

\begin{align}
  ax+by  =e \\
  cx+dy  =f
\end{align}

Para solucionarlo usaremos la regla de Cramer

  1. Calculamos el determinante
  \begin{align}
  D= ad-cb
  \end{align}
  

  2. Aplicamos la regla de Cramer

\begin{align}
   x= \frac{ed - bf}{D} \\ \\ y= \frac{af-ce}{D}
\end{align}

Con eso ya conocemos la solución del sistema

````python
import numpy as np
import matplotlib.pyplot as plt

#El usuario ingresa las variables.
a = eval(input("Escriba la variable (a) "))
b = eval(input("Escriba la variable (b) "))
e = eval(input("Escriba la variable (e) "))
c = eval(input("Escriba la variable (c) "))
d = eval(input("Escriba la variable (d) "))
f = eval(input("Escriba la variable (f) "))

#Calcula el determinante del sistema 2x2
D = (a*d) - (c*b)

#Uso de Cramer para resolver el sistema de ecuaciones
x= ((e*d)-(f*b)/D)
y= ((a*f)-(c*e)/D)

#Imprime la solución del sistem
print("la solución al sistema de ecuaciones es:", "|", "x=", x, "|", "y=", y, "|")
````

**Ejercicio 3:**
Dada las longitudes de los lados de un triángulo determine su área

---


**Solución:**
Para hallar el área se va a hacer uso de la fórmula de Herón:

\begin{align}
A = \sqrt{s(s-a)(s-b)(s-c)}
\end{align}

Donde "s" es el semiperimetro y se calcula de la siguiente forma:

\begin{align}
s= \frac{a + b + c}{2}
\end{align}

````python
import numpy as np
import matplotlib.pyplot as plt

#El usuario ingresa las variables.
a = eval(input("Digite la longitud de lado (a) sin unidades: "))
b = eval(input("Digite la longitud de lado (b) sin unidades: "))
c = eval(input("Digite la longitud de lado (c) sin unidades: "))

#Se calcula el semiperimetro
s= (a+b+c)/2

#Se usa la fórmula de Herón
A = np.sqrt(s*(s-a)*(s-b)*(s-c))

#Se imprime el área
print("El área del triángulo es:", A)
````

**Ejercicio 2:**
Dada las longitudes de los lados de un triángulo determine su área, solo si es un triángulo

---


**Solución:**
Para saber si es un triángulo o no, podemos calcular la suma de dos de sus lados, esta suma debe de ser mayor al lado restante:

\begin{align}
a+b > c \\
a+c > b \\
b+c > a \\
\end{align}

Si alguna de las anteriores no se cumple, entonces no es un triángulo y no se sigue el proceso. Caso contrario...

Para hallar el área se va a hacer uso de la fórmula de Herón:

\begin{align}
A = \sqrt{s(s-a)(s-b)(s-c)}
\end{align}

Donde "s" es el semiperimetro y se calcula de la siguiente forma:

\begin{align}
s= \frac{a + b + c}{2}
\end{align}

Y los lados del triángulo son "a", "b" y "c"

````python
import numpy as np
import matplotlib.pyplot as plt

#El usuario ingresa las variables.
a = eval(input("Digite la longitud de lado (a) sin unidades: "))
b = eval(input("Digite la longitud de lado (b) sin unidades: "))
c = eval(input("Digite la longitud de lado (c) sin unidades: "))
print()
#Se verifica que sea un triángulo
if (a+b > c) and (a+c > b) and (b+c > a):
  #Se calcula el semiperimetro
  s= (a+b+c)/2

  #Se usa la fórmula de Herón
  A = np.sqrt(s*(s-a)*(s-b)*(s-c))

  #Se imprime el área
  print("El área del triángulo es:", A, "unidades")
else: #Si no es un triángulo se salta el proceso anterior
  print("No es un triángulo")
````

````python
# Example file showing a circle moving on screen
import pygame

# pygame setup
pygame.init()
screen = pygame.display.set_mode((1280, 720))
clock = pygame.time.Clock()
running = True
dt = 0

player_pos = pygame.Vector2(screen.get_width() / 2, screen.get_height() / 2)

while running:
    # poll for events
    # pygame.QUIT event means the user clicked X to close your window
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # fill the screen with a color to wipe away anything from last frame
    screen.fill("purple")

    pygame.draw.circle(screen, "red", player_pos, 40)

    keys = pygame.key.get_pressed()
    if keys[pygame.K_w]:
        player_pos.y -= 300 * dt
    if keys[pygame.K_s]:
        player_pos.y += 300 * dt
    if keys[pygame.K_a]:
        player_pos.x -= 300 * dt
    if keys[pygame.K_d]:
        player_pos.x += 300 * dt

    # flip() the display to put your work on screen
    pygame.display.flip()

    # limits FPS to 60
    # dt is delta time in seconds since last frame, used for framerate-
    # independent physics.
    dt = clock.tick(60) / 1000

pygame.quit()
````

````python
# Método de Bisección
def biseccion(f, a, b, tol=1e-5):
  """
  Método de bisección para calcular una raiz de una función
  Datos entrada:
  f (función) Función dada  f(x)
  a , b (float) Extremos del intervalo
  tol (float) Error de aproximación, valor por defecto 1e-5
  Datos de Salida:
  m (float) Aproximación de la raiz de f
  Imprime en pantalla los resultado de cada iteración
  """

  fa = f(a)
  fb = f(b)
  k = 0
  error=abs(b-a)
  print('\n   n       a         b         f(a)        f(b)         m        f(m)     error ')
  print('----------------------------------------------------------------------------------')
  while error > tol:
    m = (a+b)/2.
    fm = f(m)
    print('{:4d} {:10.4f} {:10.4f} {:10.4f} {:10.4f} {:10.4f} {:10.4f} {:10.4f}' .format(k, a, b, fa, fb, m, fm, error))
    if (fm == 0.0):
      return m
    if fa*fm < 0.0:    # nuevo intervalo [a,m]
       b = m
       fb = fm
    else:             # nuevo intervalo [m,b]
       a = m
       fa = fm
    error=abs(b-a)
    k = k+1
  if abs(fa) < abs(fb):
     return a
  else:
     return b
````

**Iteración de punto fijo**

(Diagrama de telaraña)  


````python
import math as m

def puntofijo(x0, xk, g, tol = 1e-2):
  """
  Esta es una función que calcula el punto fijo.....

  """
  gx = g(x0)#x0 + 1 - (x0 * m.e**x0) #(1+x0)/(1+m.e**x0) #m.e**-x0
  error = abs(xk-x0)
  n = 0

  print('   n       x0         xk        g(x)       error ')
  print('---------------------------------------------------')
  while error > tol:
    print('{:4d} {:10.4f} {:10.4f} {:10.4f} {:10.4f}' .format(n, x0, xk, gx, error))
    n = n+1
    gx = g(x0)#x0 + 1 - (x0 * m.e**x0)  #(1+x0)/(1+m.e**x0) #m.e**-x0
    xk = gx # xk = g(x)
    error = abs(xk-x0)
    x0 = xk # Cambia el valor de x0
````

````python
import math as m

def newton(x0, xk, g, tol = 1e-16):
  """
  Esta es una función que calcula el punto fijo por método de newton.....

  """
  gx = ((x0**2)*(m.e**x0)+1)/((m.e**x0)*(1+x0))
  error = abs(xk-x0)
  n = 0

  print('   n       x0         xk        g(x)       error ')
  print('---------------------------------------------------')
  while error > tol:
    print('{:4d} {:10.4f} {:10.4f} {:10.4f} {:10.4f}' .format(n, x0, xk, gx, error))
    n = n+1
    gx = ((x0**2)*(m.e**x0)+1)/((m.e**x0)*(1+x0))
    xk = gx # xk = g(x)
    error = abs(xk-x0)
    x0 = xk # Cambia el valor de x0
````

````python
import numpy as np

def Gauss(A,b):
  """
  Soluciona A*x=b mediante eliminación de Gauss
  """
  A0 = A.copy
  b0 = b.copy
  n = len(b)
  x = np.zeros((n,1))
  for k in range (n-1): #Eliminación
    for i in range (k+1,n):
      if A[i,k] != 0.0:
       mult = A[i,k] / A[k,k]
       A[i,k+1:n] = A[i,k+1:n] - mult* A[k,k+1:n]
       b[i] = b[i] - mult * b[k]
       #print("°",k,A)
  for k in range(n-1,-1,-1):
    x[k] = (b[k] - np.dot(A[k,k+1:n],x[k+1:n]))/A[k,k] #A@x - b
    #print(A@x -b)
  return x
````

````python
def FactLU(A):
  """
  Factorización LU de una matriz A
  Datos entradas
    A : Matriz nxn
  Datos salidas
   A contiene las matrices L y U
  """
  n = len(A)
  for k in range (0,n-1):
    for i in range(k+1,n):
      if A[i,k] != 0.0:
        mul= A[i,k]/A[k,k]
        A[i,k+1:n] = A[i,k+1:n] - mul * A[k,k+1:n]
        A[i,k] = mul #Se guardan los valores de L
  return A
````

````python
def solucionLU(A,b):
  """
  solución LU de una matriz A
  Datos entrada
    A : [L/U] matriz que contiene la factorización LU
    b : vector de tamaño nx1
  Datos salida
    x = solucion(A,b)
    resuelve LU x = b
  """
  n = len(A)
  for k in range(1,n):
    b[k] = b[k] - np.dot(A[k,0:k],b[0:k])
  b[n-1] = b[n-1]/A[n-1,n-1]

  for k in range(n-2,-1,-1):
    b[k] = (b[k] - np.dot(A[k,k+1:n], b[k+1:1])) / A[k,k]
  return b
````

````python
def cambiofilas(v,i,j):
  """
  Cambia la fila i y la fila k del vector o matriz [v]
  """
  if len(v.shape) == 1:
    v[i],v[j] = v[j],v[i]

  else:
    v[[i,j],:] = v[[j,i],:]
  return v
````

````python
def gaussPivot(a,b,tol=1.0e-12):
  """
  x = gaussPivot(a,b,tol=1.0e-12)
  Soluciona [a]{x} = {b} mediante eliminación de Gauss
  con pivoteo parcial
  """

  n = len(b)
  x = np.zeros((n,1))
  #Busca el pivote de mayor magnitud
  s = np.zeros(n)
  for i in range(n):
    s[i] = max(np.abs(a[i,:]))

  #pt 2
  for k in range(0,n-1):
    #Intercambio de filas, si es necesario
    p = np.argmax(np.abs(a[k:n,k])/s[k:n]) + k
    if abs(a[p,k]) < tol:
      print("la matriz es singular")
    if p!= k:
      cambiofilas(b,k,p)
      cambiofilas(s,k,p)
      cambiofilas(a,k,p)
    #Eliminación
    for i in range(k+1,n):
      if a[i,k] != 0.0:
        mult = a[i,k]/a[k,k]
        a[i,k+1:n] = a[i,k+1:n] - mult * a[k,k+1:n]
        b[i] = b[i] - mult * b[k]
    return a
````

````python
#Programa creado por el profesor


def gaussPivot(A, b):
    """
    x = gaussPivot(A,b).
    Soluciona Ax = b mediante la eleminación de Gauss con pivoteo parcial
    equivalente a la factorizacion PA = LU
    """
    N, M = A.shape
    X = np.zeros((N, 1))
    Y = np.zeros((N, 1))
    C = np.zeros((1, N))  # var auxiliar
    R = np.arange(0, N )  # Vector de indices

    for p in range(N):
        # Identifica el elemento Pivote
        max1, j = np.max(np.abs(A[p:N, p])), np.argmax(np.abs(A[p:N, p])) + p

        # Intercambiar filas p y j
        C = A[p, :].copy()
        A[p, :] = A[j, :].copy()
        A[j, :] = C.copy()
        d = R[p]
        R[p] = R[j]
        R[j] = d

        if A[p, p] ==0: # 1e-12:
            print(A[p,p])
            print("A es singular. No hay una solución única")
            return

        # Eiminación de las componetes pore debaj de la diagonal
        for k in range(p + 1, N):
            mult = A[k, p] / A[p, p]
            A[k, p] = mult
            A[k, p + 1:N] = A[k, p + 1:N] - mult * A[p, p + 1:N]

    Y[0] = b[R[0]] # Se reorganiza el b según los cambios de fila

    #Solucion sistema triangular inferior
    for k in range(1, N):
        Y[k] = b[R[k]] - np.dot(A[k, :k], Y[:k])

    #Solucion sistema triangular superior
    X[N - 1] = Y[N - 1] / A[N - 1, N - 1]
    for k in range(N-2,-1,-1):
        X[k] = (Y[k] - np.dot(A[k, k + 1:N], X[k + 1:N])) / A[k, k]

    return X
````

````python
import numpy as np
import matplotlib.pyplot as plt

def trapecio(f, a, b, n):
  """
  f = función
  a = inicio del intervalo
  b = fin del intervalo
  n = número de subintervalos
  """
  F = lambda x : eval(str(f))
  #print(type(F), "|", type(f), "|", str(f))
  X=[a]
  Y = [F(a)]#[np.sqrt(1 + a**2)]
  P = 0
  h = (b-a)/n
  xi = a #Toma el primer valor del intervalo
  xii = xi + h #Se crea el primer subointervalo [xi, xi+1]


  for i in range(0,n):
    #print(xi,xii)
    xi = xii
    xii = xi + h #Se crea el subintervalo [xi, xi+1]
    x = xi
    #F = F(x) #np.sqrt(1 + x**2)
    P = F(x) + P
    #print("►►►",F(x))
    X.append(xi)
    Y.append(F(x))
  plt.axes().set_aspect('equal')
  plt.plot(X, Y, '-', color=(0.5, 0., 0.5))
  plt.grid(True)
  plt.show()
  return((h/2) * (a + b + (2*P)))
````

##Regla de Simpson: (Kepler 1620)
► Se particiona [a,b] en un número par 2n subintervalos
```
x0 = a, x2n = b,  h = (b-a)/2n,   xi+1-xi = h
```

\begin{align}
\int_{xi}^{x2i+2}fi(x) dx
\end{align}

````python
def simpson(f,a,b,n):
  h = (b-a)/n
  ximpar = np.array(a+h,b,2*h)
  xpar = np.ary(a+2*h, b-h, 2*h)
  s = f(a) + 4 * sum(f(ximpar)) + 2 * sum(f(xpar)) + f(b)
  return (h/3)*s
````

````python
class Node(Point):
  """
  Clase que hereda de la clase Point
  """
  numberNode = 0
  def __init__(self,xx,yy): #Construye un punto
    super().__init__(xx,yy) #Inclusión del atributo de identificador del nodo
    self.id = Node.numberNode #Asigna el identificador de nodo
    Node.numberNode += 1

class Bar:
  """
  Define una barra soportada por dos nodos
  """
  def __init__(self,n1,n2):
    if n1.id == n2.id: #Son el mismo punto
      print("Error: No hay barra")
      return

    elif n1.id < n2.id:
      self.orig = n1
      self.fin = n2

    else:
      self.orig = n2
      self.fin = n1

  def __str__(self):
    try:
      return "Barra de extremos los nodos {0} y {1}".format(self.orig.id, self.fin.id)
    except AttributeError:
      return ""
````

````python
class Truss :
    """
    genera una estructura rectangular de barras
    - nx: numero de nodos en abscisas .
    - ny: numero de nodos en ordenadas .
    - p1: vértice inferior izquierdo (clase punto).
    - p2: vértice superior derecho ( clase punto ).
    genera 6 barras entre 4 nodos
    """

    def __init__ (self ,p1 ,p2 ,nx ,ny): #Constructor de la estructura
        # comprobación de la integridad del rectángulo
        if p2.x-p1.x < 1.e-6 or p2.y-p1.y < 1.e-6:
            print(" Rectángulo incorrecto ")
            return

        self.nNodos = (nx + 1) * (ny + 1)
        self.nBarras = 4* nx*ny+ny+nx
        self.nodos = []
        self.barras = []

        Node. numberNode = 0

        # construcción de nodos
        nodx = np.linspace (p1.x,p2.x,nx +1)
        nody = np.linspace (p1.y,p2.y,ny +1)

        for yy in nody:
            for xx in nodx:
                self.nodos. append (Node(xx ,yy))

        # construcción de barras
        for j in range (ny):
            for i in range (nx):
                n1 = i+ j*(nx +1)
                n2 = n1 + 1
                n3 = n1 + nx + 1
                n4 = n3 + 1
                # barras en cada elemento
                b1 = Bar(self.nodos[n1],self.nodos[n2])
                b2 = Bar(self.nodos[n1],self.nodos[n3])
                b3 = Bar(self.nodos[n1],self.nodos[n4])
                b4 = Bar(self.nodos[n2],self.nodos[n3])
                self. barras . extend ([b1 ,b2 ,b3 ,b4])
                # barras finales a la derecha
                self. barras . append (Bar(self.nodos [n2],self.nodos[n4]))


        # barras de la línea superior
        indice =ny *( nx +1) +1
        for j in range (nx):
            self. barras.append (Bar(self.nodos[ indice +j-1] , self.nodos[indice+ j ]))

    # Métodos

    def get_coordinate(self):
        coordenadas =[]
        for nod in self.nodos:
            coordenadas.append ([ nod.x,nod.y])
        return np.array( coordenadas )

    def get_connection(self):
        conexiones =[]
        for bar in self.barras :
            conexiones.append ([ bar.orig.id ,bar.fin.id ])
        return np.array( conexiones )
        conexiones =[]
        for bar in self.barras :
            conexiones.append ([ bar.orig.id ,bar.fin.id ])
        return np.array( conexiones )


    # Dibujar la estructura

    def plotting( self ):
        plt.ion()
        fig = plt.figure ()
        bx = fig.add_subplot (111)

        for bar in self .barras:
            bx.plot([bar.orig.x,bar.fin.x],[bar.orig.y,bar
            .fin.y], 'D-' , linewidth =2)
        bx.axis( 'equal' )
        plt.show()

    # Método eliminar una barra
    def remove_bar(self ,n1 ,n2):
        if n2 < n1:
            n1, n2 = n2, n1
        elif n1 == n2:
            print("Nodos incorrectos ")

        for bar in self.barras :
            if bar.orig.id == n1 and bar.fin.id == n2:
                self.barras.remove(bar)
                self.nBarras -= 1
                return
            else:
                print("No existe tal barra ")

````

####Método de Runge-Kutta
\begin{align}
s_1 = f(x_k, t_k)
\end{align}

\begin{align}
s_2 = f(x_k + \frac{h}{2}*s_1, t_k + \frac{h}{2})  
\end{align}

\begin{align}
s_3 = f(x_k + \frac{h}{2}*s_2, t_k + \frac{h}{2})
\end{align}

\begin{align}
s_4 = f(x_k *s_3, t_k + h)
\end{align}

\begin{align}
x_{k+1} = x_k + \frac{h}{6} [s_1 + 2*s_2 + 2*s_3 + s_4]
\end{align}

````python
import numpy as np
import matplotlib.pyplot as plt

class OneStepMethod:
  def __init__(self, f, x0, interval, N):
    self.f = f
    self.x0 = x0
    self.interval = [t0, te] = interval
    self.grid = np.linspace(t0, te, N)
    self.h = (te - t0) / N

  def generate(self):
    ti, ui = self.grid[0], self.x0
    yield ti, ui
    for t in self.grid[1:]:
      ui = ui + self.h * self.step(self.f, ui, ti)
      ti = t
      yield ti, ui

  def solve(self):
    self.solution = np.array(list(self.generate()))

  def plot(self):
    plt.plot(self.solution[:,0], self.solution[:,1])

  def print_solution(self):
    print("ti\tui")
    for ti, ui in self.solution:
      print(f"{ti:.6f}\t{ui:.6f}")

class EulerExplicito(OneStepMethod):
  def step(self, f, u, t):
    return f(u, t)

class ReglaPuntoMedio(OneStepMethod):
  def step(self, f, u, t):
    s1 = f(u,t)
    return f(u + self.h/2 * s1, t + self.h/2)

class RungeKutta4(OneStepMethod):
  def step(self, f, u, t):
    s1 = f(u, t)
    s2 = f(u + self.h/2 * s1, t + self.h/2)
    s3 = f(u + self.h/2 * s2, t + self.h/2 )
    s4 = f(u + s3, t + self.h)
    return (1/6)* (s1 + 2*s2 + 2*s3 + s4)
````

**Cálculo del máximo comun divisor**

Consideremos dos enteros positivos $p$ y $q$. Su máximo común divisor se puede calcular por el algoritmo de Euclides. Este es un proceso iterativo en el que en cada iteración, dados dos enteros positivos $p_k \ge q_k$, se halla el cociente $c_k$ y el residuo $r_k$ al dividir $p_k$ por $q_k$:
$$
p_k = c_k q_k + r_k
$$
Para pasar a la nueva iteración, el divisor pasa a ser dividendo y el residuo pasa a ser divisor.
\begin{align*}
p_{k+1} = q_k, \\
q_{k+1} = r_k·
\end{align*}
El proceso se repite hasta que el residuo sea nulo. El máximo común
divisor estará dado por el último divisor $q$. Por ejemplo, consideremos
$p = 60$, $q =18$. La siguiente tabla muestra los pasos intermedios del
algoritmo
$$
\begin{array}[rrrr]
p_k &  q_k & c_k &  r_k \\
60 &  18 & 3 &  6 \\
185 &  6 & 3 &   0 \\
\end{array}
$$
Entonces el máximo común divisor es 6.


Para $p = 34$, $q =15$. La siguiente tabla muestra los pasos intermedios del
algoritmo
$$
\begin{array}[rrrr]
p_k &  q_k & c_k &  r_k \\
34 &  15 & 2 &  4 \\
15 &  4 & 3 &   3 \\
4 &  3 &  1 &  1 \\
3 &  1 &  3&  0
\end{array}
$$
Entonces el máximo común divisor es 1, es decir, 34 y 15 son primos
relativos.

````python
import numpy as np

###################################################################################################
#---------------------------###### Simplificar racionales ######----------------------------------#
###################################################################################################
def simplificar(a,b):
  """
  Simplifica la razón entre dos enteros
  • Datos de entrada:
    ► [ a ] : Numerador de la fracción
    ► [ b ] : Denominador de la fracción

  • Datos de salida:
    ► [ c/d ] : Fracción simplificada
  """
  print()
  if b == 0:
    return("No es una fracción valida")
  if a == 0:
    print(0)
    return
  a0,b0 = a, b #Crea una copia de los numeros introducidos
  if a0 < 0 and b0 <0:
    a0, b0 = -a0, -b0


  #En caso de que se trate de un número negativo se vuelve positivo para realizar la división
  if a < 0:
    a = -a
  if b < 0:
    b = -b
  ##- Se verifica que p sea mayor que q, si no es el caso se intercambian -##
  if a<b:
    a, b = b, a

  c = a//b #/Halla el cociente entre p y q
  r = a%b #/Halla el residuo entre p y q

  ##- Crea un bucle que se ejecuta mientras el residuo (r) sea mayor o igual que 0
  while r>=0:
    if r==0: #/Termina el bucle cuando el residuo es cero
      break
    a, b = b, r #/Intercambia las variables para volver a realizar los cálculos de residuo y cociente
    c, r = a//b, a%b
  if b0/b == 1:
    print(a0/b)
  else:
    print(a0/b, "/", b0/b)

###################################################################################################
#---------###### Operaciones (Suma,Resta,Multiplicación,División) entre racionales ######---------#
###################################################################################################
def operaciones(a, b, c, d):
  """
  Realiza las 4 funciones elementales entre dos racionales
  • Datos de entrada:
    ► [ a/b ] : Racional de la primer posición
    ► [ c/d ] : Racional de la segunda posición

  • Datos de salida:
    ► [ Suma ]
    ► [ Resta ]
    ► [ Multiplicación ]
    ► [ División ]
  """

  #Se definen operaciones sencillas que se repiten en las operaciones
  ad = a*d
  bc = b*c
  ac = a*c
  bd = b*d

  def Suma():
    numerador = (ad)+(bc)
    denominador = bd
    print("\n La suma es: ", end="")
    simplificar(numerador, denominador)
  def Resta():
    numerador = (ad)-(bc)
    denominador = bd
    print("\n La resta es: ",  end="")
    simplificar(numerador, denominador)
  def Multiplicacion():
    numerador = ac
    denominador = bd
    print("\n La multiplicación es: ",  end="")
    simplificar(numerador, denominador)
  def Division():
    numerador = ad
    denominador = bc
    print("\n La división es: ", end="")
    simplificar(numerador, denominador)

  print("\n" ,Suma(), Resta(), Multiplicacion() ,Division())

###################################################################################################
#-----------------------------###### Polinomio de Legendre ######---------------------------------#
###################################################################################################
def polL(x, n):
  """
  Calcula el polinomio de Legendre de grado "n" para un "x" dado, \ siendo "n" un entero no negativo
  • Datos de entrada:
    ► [ x ] : Número a evaluar
    ► [ n ] : El grado del polinomio

  • Datos de salida:
    ► [ Polinomio de Legrenge de grado n evaluado en x ]
  """
  L0 = 1
  L1 = x
  for i in range(1, n):
    D = (2*i) + 1
    N = (i+1)
    L = D/N * (x * L1 - (i / D) * L0)
    L0, L1 = L1, L
    i = i+1
  print(L)

###################################################################################################
#-----------------------------###### Método de Bisección ######-----------------------------------#
###################################################################################################
def biseccion(f, a, b, tol=1e-5): # Método de Bisección creado por el profesor con pequeñas modificaciones mías
  """
  Método de bisección para calcular una raiz de una función dada
  • Datos entrada:
    ► [ f ] : Función dada
    ► [ a ] : Primer extremo del intervalo a evaluar
    ► [ b ] : Segundo extremo del intervalor a evaluar

  • Datos de Salida:
    ► [ x! ] : Aproximación a la raiz de f
    ► [ Biseccion.dat ] : Archivo con las iteraciones para hallar la estimación de la raiz de la función
  """

  fa = f(a)
  fb = f(b)
  n = 1
  error=abs(b-a)
  with open("Biseccion.dat", "w") as B:
    B.write(u'   n        x\u2099         ƒ(x\u2099)       Error  \n')
    B.write('-------------------------------------------\n')
    while error > tol:
      x = (a+b)/2.
      fx = f(x)
      B.write('{:4d} {:10.5f} {:10.5f} {:10.5f}' .format(n, x, fx, error)+"\n")
      if (fx == 0.0):
        return x
      if fa*fx < 0.0:    # nuevo intervalo [a,x]
        b = x
        fb = fx
      else:             # nuevo intervalo [m,x]
        a = x
        fa = fx
      error=abs(b-a)
      n = n+1
    if abs(fa) < abs(fb):
      return a
    else:
      return b

###################################################################################################
#-----------------------------###### Método de Punto Fijo ######----------------------------------#
###################################################################################################
def puntofijo(g, x, tol = 1e-5):
  """
  Calcula una raiz de una funcion g mediante el uso del método de Punto Fijo, donde g(x) = -x
  • Datos entrada:
    ► [ g ] : Función dada
    ► [ x0 ] : Punto desde donde se quiere empezar a encontrar la raíz

  • Datos de Salida:
    ► [ x! ] : Aproximación a la raiz de f
    ► [ PuntoFijo.dat ] : Archivo con las iteraciones para hallar la estimación de la raiz de la función
  """
  gx = g(x)
  error = abs(0-x)
  n = 0
  with open("PuntoFijo.dat", "w") as P:
    P.write(u'   n        x\u2099         ƒ(x\u2099)       Error  \n')
    P.write('-------------------------------------------\n')
    P.write('{:4d} {:10.5f} {:10.5f} {:10.5f}' .format(n, x, gx, error)+"\n")
    while error > tol:
      xk = gx
      n = n+1
      gx =  g(xk)
      error = abs(xk-x)
      x = xk
      P.write('{:4d} {:10.4f} {:10.4f} {:10.4f}' .format(n, x, gx, error)+"\n")
      if n>5000:
        print("Superado el límite de iteraciones en PuntoFijo, la funcion diverge")
        break
    return(xk)

###################################################################################################
#----------------------------###### Derivada de una función ######--------------------------------#
###################################################################################################
def derivada(f,x,h):
  """
  Es una función que calcula la derivada de una función en un x dado

  • Datos entrada:
    ► [ f ] : Función
    ► [ x ] : X a evaluar
    ► [ h ] : Número muy cercano a 0

  • Datos salida:
    ► [ g ] : Derivada evaluada en x
  """
  fx = f(x)
  xh = x+h
  fh = f(xh)
  g = (fh - fx) / h
  return(g)

###################################################################################################
#-------------------------------###### Método de Newton ######------------------------------------#
###################################################################################################
def newton(g, x0, xk, tol = 1e-5):
  """
  Esta es una función que encuentra una raiz de una función por el método de newton
  • Datos entrada:
    ► [ g ] : Función dada
    ► [ x0 ] : Punto desde donde se quiere empezar a encontrar la raíz
    ► [ xk ] : Punto para calcular el error

  • Datos de Salida:
    ► [ x! ] : Aproximación a la raiz de f
    ► [ Newton.dat ] : Archivo con las iteraciones para hallar la estimación de la raiz de la función
  """
  gx = x0 - (g(x0) / derivada(g, x0, 0.0001))
  error = abs(xk-x0)
  n = 0

  with open("Newton.dat", "w") as N:
    N.write(u'   n        x\u2099         ƒ(x\u2099)       Error  \n')
    N.write('-------------------------------------------\n')
    while error > tol:
      N.write('{:4d} {:10.5f} {:10.5f} {:10.5f}' .format(n, x0, gx, error)+"\n")
      n = n+1
      gx = x0 - (g(x0) / derivada(g, x0, 0.0001))
      xk = gx # xk = g(x)
      error = abs(xk-x0)
      x0 = xk # Cambia el valor de x0
      if n>5000:
        print("Superado el límite de iteraciones, la funcion diverge")
        break
    return(x0)

###################################################################################################
#-----------------------###### Raices de ecuaciones no lineales ######----------------------------#
###################################################################################################
def ecuacion_no_lineal(f, a, b):
  """
  Halla raices a ecuaciones no lineales por medio de los métodos de:
  Bisección, Punto fijo y Newton.

  • Datos de entrada:
    ► [ F ] : Función en Numpy(np.función)
    ► [ a ] : Primer extremo del intervalo a evaluar
    ► [ b ] : Segundo extremo del intervalor a evaluar

  • Datos de salida:
    ► [ Newton.dat ] : Tabla con iteraciones del método de Newton
    ► [ Biseccion.dat ] : Tabla con iteraciones del método de Bisección
    ► [ PuntoFijo.dat ] : Tabla con iteraciones del método de Punto Fijo

  """
  print("Por el método de Newton la raiz aproximada es: ", newton(f, a, b))
  print("y su tabla de iteraciones es la siguiente: \n")
  with open("Newton.dat", "r") as n:
    print(n.read())
  print()

  print("Por el método de bisección la raiz aproximada es: ", biseccion(f, a, b))
  print("y su tabla de iteraciones es la siguiente: \n")
  with open("Biseccion.dat", "r") as b:
    print(b.read())
  print()
  print("Por el método de PuntoFijo la raiz aproximada es: ", puntofijo(f, a))
  print("y su tabla de iteraciones es la siguiente: \n")
  with open("PuntoFijo.dat", "r") as p:
    print(p.read())

###################################################################################################
#-------------------------###### Punto Mínimo función unimodal ######-----------------------------#
###################################################################################################
def unimodal(x1,s):
  """
  Encuentra el punto mínimo o un número muy cercano a este
  De una función unimodal

  • Datos de entrada:
    ► [ x1 ] : Punto inicial
    ► [ s ] : Tamaño del peso

  • Datos de salida:
    ► [ x! ] : Punto minimo estimado
    ► [ FuncionUnimodal.dat ] : Archivo con las iteraciones para hallar la estimación del punto mínimo
  """
  paso = s
  x2 = x1 + paso

  def funcion(x):
    f = x * (x - 1.5) # la función se puede reemplazar por otra que sea unimodal, y seguirá encontrando un punto cercano al punto mínimo de dicha función
    return(f)

  f1 = funcion(x1)
  f2 = funcion(x2)
  l = 1
  i = 0
  with open("FuncionUnimodal.dat", "w") as U:
    U.write(u'   i        s        x\u1D62 = x\u1D62\u208B\u2081 + s        f\u1D62 = f(x\u1D62) \n')
    U.write('---------------------------------------------------------\n')
    while l>0:
      if f2<f1:
        U.write('{:4d} {:10.2f} {:13.2f} {:22.8f}' .format(i, paso, x1, f1)+"\n")
        x1 = x2
        x2 = x1 + paso
        f1 = funcion(x1)
        f2 = funcion(x2)
        i =  i + 1
        if f2>f1:
          U.write('{:4d} {:10.2f} {:13.2f} {:22.8f}' .format(i, paso, x1, f1)+"\n")
          print("\nEl punto mínimo estimado es: ", f1)
          break

      if f2>f1:
        U.write('{:4d} {:10.2f} {:13.2f} {:22.8f}' .format(i, paso, x1, f1)+"\n")
        x2 = x1
        x1 = x1 - paso
        f1 = funcion(x1)
        f2 = funcion(x2)
        i = i - 1
        if f2<f1:
          print("\nEl punto mínimo estimado es: ", f2)
          break

      if f2==f1:
        U.write('{:4d} {:10.2f} {:13.2f} {:22.8f}' .format(i, paso, x1, f1)+"\n")
        print("\nEl punto mínimo estimado es: ", f1)
        break

###################################################################################################
#--------------------------------###### Producto punto ######-------------------------------------#
###################################################################################################
def p_punto(v1,v2):
  """
  Realiza el producto punto entre 2 vectores tranformandolos en matrices
  columna y fila, para realizar el producto entre matrices

  • Datos de entrada:
    ► [ v1 ] : Primer vector
    ► [ v2 ] : Segundo vector

  • Datos de salida:
    ► [ R ] : Producto punto
  """

  if type(v1) == int: #Verifica si es un vector de Dimension 1, si lo es entonces n1 = 1
    n1 = 1
  else:
    n1 = len(v1) #En caso de que el vector sea de Dimensión superior a 1, entonces n1 = Dimension del vector
  if type(v2) == int:
    n2 = 1
  else:
    n2 = len(v2)

  if n1 != n2:
    print("Dos vectores de distintas dimensiones no se pueden multiplicar") #En producto punto se deben de hacer operaciones componente a componente, si no tienen la misma cantidad no se realiza el proceso
    return
  v1 = np.array(v1).reshape(n1,1) #Transpone el vector y lo transforma en matriz columna
  v2 = np.array(v2).reshape(1,n2) #Transforma el vector en matriz fila

  print("El producto punto del vector [v1] =", np.array(v1).reshape(1,n1), "y del vector [v2] =", v2, "es: ", np.dot(v2,v1))

###################################################################################################
#-------------------------------###### Suma de Matrices ######------------------------------------#
###################################################################################################
def s_matrices(m1,c1,f1,m2,c2,f2,SR):
  """
  Calcula la suma (o resta) entre dos matrices del mismo tamaño

  • Datos de entrada:
    ► [ m1 ] : Primer matriz
    ► [ c1 ] : Cantidad de columnas de la primer matriz
    ► [ f1 ] : Cantidad de filas de la primer matriz
    ► [ m2 ] : Segunda matriz
    ► [ c2 ] : Cantidad de columnas de la segunda matriz
    ► [ f2 ] : Cantidad de filas de la segunda matriz
    ► [ S/R ] : Modo de suma o resta

  • Datos de salida:
    ► [ S/R ] :  Suma/Resta entre las dos matrices
  """
  if c1 != c2:
    print("Sus matrices no tienen la misma cantidad de columnas")
    return
  if f1 != f2:
    print("Sus matrices no tienen la misma cantidad de filas")
    return

  if np.size(m1) != c1*f1:
    print("La cantidad de componentes de su primer matriz no encajan en las filas y columnas dadas")
    return

  if np.size(m2) != c2*f2:
    print("La cantidad de componentes de su segunda matriz no encajan en las filas y columnas dadas")
    return

  m1 = np.array((m1)).reshape(c1,f1)
  m2 = np.array((m2)).reshape(c2,f2)
  if SR == 0:
    print("El resultado de la suma es:\n",m1+m2)
  elif SR == 1:
    print("El resultado de la resta es:\n",m1-m2)
  else:
    print("No se seleccionó modo de Suma ni de Resta")

###################################################################################################
#--------------------------###### Multiplicación de Matrices ######-------------------------------#
###################################################################################################
def m_matrices(A,c1,f1,B,c2,f2):
  """
  Multiplica dos matrices del Modo A*B donde la cantidad de Columnas de A es la misma
  cantidad de Filas que tiene B.

  • Datos de entrada:
    ► [ A ] : Primer matriz
    ► [ c1 ] : Cantidad de columnas de la primer matriz
    ► [ f1 ] : Cantidad de filas de la primer matriz
    ► [ B ] : Segunda matriz
    ► [ c2 ] : Cantidad de columnas de la segunda matriz
    ► [ f2 ] : Cantidad de filas de la segunda matriz

  • Datos de salida:
    ► [ M ] :  Resultado de la multiplicación entre las dos matrices
  """

  if c1 != f2:
    print("Su primer matriz no tiene la misma cantidad de Columnas como Filas que tiene su segunda matriz")
    return

  if np.size(A) != c1*f1:
    print("La cantidad de componentes de su primer matriz no encajan en las filas y columnas dadas")
    return

  if np.size(B) != c2*f2:
    print("La cantidad de componentes de su segunda matriz no encajan en las filas y columnas dadas")
    return

  A = np.array((A)).reshape(f1,c1)
  B = np.array((B)).reshape(f2,c2)

  print("El resultado de la multiplicación es:\n", A@B)

###################################################################################################
````

**Definición:**  Una función es unimodal si

  (i) $x_1 < x_2 < x^*$ implica que $f(x_2)<f(x_1)$, y

  (ii) $x_2 > x_1 > x^*$ implica que $f(x_1)< f(x_2)$,

donde $x^*$ es el punto mínimo


El método más elemental para determinar el mí nimo o máximo de una función unimodal es el método de busqueda de paso fijo. Este método se describe por los siguientes pasos:

1. Inicie con un punto  inicial $x_1$
2. Encuentre $f_1 = f(x_1)$.
3. Suponga un tamaño de paso $s$, defina $x_2=x_1+s$.
4. Encuentre $f_2 = f(x_2)$.
5. Si $f_2 < f_1$ y si el problema es de minimización, dado que la función es unimodal el mínimo no puede ser un $x$ tal
que $x < x_!$. Por lo que la busqueda continúa por los puntos
$x_3, x_4, ...$ usando la propiedad de unimodlaidad mientras se
verifica cada par de evaluaciones de la función. Este proceso
continúa hasta un punto $x_i = x_1 + (i-1)S $ que muestra un
incremento de la función.
6. La busqueda se termina en $x_i$, y puede considerarse como míninmo $x_i-1$ o
$x_i$.
7. Originalmente si $f_2 > f_1$ la busqueda debe llevarse a
cabo en dirección contraria en los puntos  $x_{-2}, x_{-3},...$ donde $x_j= x_i - (j-1)s$.
8. Si $f_2=f_1$, el mínimo estáa entre $x_1$ y $x_2$. El
mínimo puede considerase como $x_1$ o $x_2$.
\item Si ocurre que ambos $f_{-2}$ y $f_2$ son mayores que $f_1$,
esto implica que el mínimo estará en el intervalo $x_{-2} <
x < x_2$.
